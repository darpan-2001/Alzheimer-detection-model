```python
cnn_model = Sequential()

cnn_model.add(Conv2D(filters=32, kernel_size=(3,3), activation='relu', input_shape=(220,220,3)))
cnn_model.add(MaxPooling2D(2,2))
cnn_model.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu'))
cnn_model.add(MaxPooling2D(2,2))
cnn_model.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu'))
cnn_model.add(MaxPooling2D(2,2))
cnn_model.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu'))
cnn_model.add(MaxPooling2D(2,2))

cnn_model.add(Flatten())

cnn_model.add(Dense(512, activation='relu'))
cnn_model.add(Dropout(0.5))
cnn_model.add(Dense(4, activation='softmax'))
```

```python
cnn_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
cnn_model.summary()
```

```python
data_gen = ImageDataGenerator(rescale=1/255)
```

```python
training_data = data_gen.flow_from_directory('DATA/train', target_size=(220,220), class_mode='categorical', batch_size=32)
```

```python
testing_data = data_gen.flow_from_directory('DATA/test', target_size=(220,220), class_mode='categorical', batch_size=32)
```

```python
training_data.class_indices
```

```python
testing_data.class_indices
```

```python
cnn_model.fit(training_data, steps_per_epoch=len(training_data), epochs=5)
```
